```python
df = pd.read_csv('Morticd9.csv', dtype={'2':np.int32})
```

```python
df = df.drop(['Admin1', 
              'SubDiv', 
              'Frmat', 
              'IM_Frmat', 
              'Deaths1', 
              'Deaths2',
              'Deaths3',
              'Deaths4',
              'Deaths5',
              'Deaths6',
              'Deaths7',
              'Deaths8',
              'Deaths9',
              'IM_Deaths1', 
              'IM_Deaths2', 
              'IM_Deaths3', 
              'IM_Deaths4'], 
               axis=1)
```

```python
df = pd.melt(df, 
             id_vars=['Country',
                      'Year',
                      'List',
                      'Cause',
                      'Sex'], 
             var_name='Age',
             value_name='Freq')
```

```python
print(df.shape)
```

```python
df = df.replace(0, pd.np.nan).dropna()
```

```python
allCauses = df[df['Cause'] != 'B00'].index

df.drop(allCauses, inplace=True)
```

```python
print(df.shape)
```

```python
df_1 = df.groupby(['List', 'Cause', 'Age'], as_index=False).agg({'Freq':'sum'})
```

```python
print(df_1.shape)
```

```python
df_1.head(10)
```

```python
df_1.tail(10)
```

```python
sparseCauses = df_1[df_1['Freq'] < 1000000].index

df_1.drop(sparseCauses, inplace=True)

print(df_1.shape)
```

```python
sns.barplot(x='Age', y='Freq', data=df_1, ci=None)
plt.show()
```

```python
df_1.to_csv('cleaned_df_1.csv', index=None)
```
