```python
# Import libraries

import numpy as np

import pandas as pd
pd.set_option('display.max_columns', 100)

from matplotlib import pyplot as plt
%matplotlib inline 

import seaborn as sns
sns.set_style('darkgrid')
```

```python
# Read dataset

df = pd.read_csv('cancer-deaths-by-age.csv')

df.head(10)
```

```python
print(df.shape)
```

```python
# Convert to integer

df = df.astype({'Under-5s (deaths)':'int64',
                '5-14 years old (deaths)':'int64',
                '15-49 years old (deaths)':'int64',
                '70+ years old (deaths)':'int64',
                '50-69 years old (deaths)':'int64'})
```

```python
df.dtypes
```

```python
df.head(10)
```

```python
# Drop 'Entity' and 'Code' columns, group by year, aggregate age columns, assign to a new dataframe

df_1 = df.groupby(['Year'], as_index=False).agg({'Under-5s (deaths)':'sum', 
                                                 '5-14 years old (deaths)':'sum', 
                                                 '15-49 years old (deaths)':'sum', 
                                                 '70+ years old (deaths)':'sum', 
                                                 '50-69 years old (deaths)':'sum'})

print(df_1.shape)
```

```python
df_1.head(10)
```

```python
# Use melt() to combine age features into one column, 'Age' 
# Create new column, 'Total'  

df_1 = pd.melt(df_1, 
               id_vars=['Year'], 
               var_name='Age',
               value_name='Total')

df_1.head(10)
```

```python
print(df_1.shape)
```

```python
# Plot result

chart = sns.barplot(x='Age', y='Total', data=df_1, ci=None)

chart.set_xticklabels(chart.get_xticklabels(), rotation=45)

plt.show()
```

```python
# Save cleaned dataframe as a new csv

df_1.to_csv('cleaned_cancer_deaths_by_aging.csv', index=None)
```
