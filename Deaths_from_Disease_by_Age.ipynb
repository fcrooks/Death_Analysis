```python
# Import libraries

import numpy as np

import pandas as pd
pd.set_option('display.max_columns', 100)

from matplotlib import pyplot as plt
%matplotlib inline 

import seaborn as sns
sns.set_style('darkgrid')
```

```python
# Read dataset

df = pd.read_csv('Morticd9.csv', dtype={'2':np.int32})
```

```python
# Drop irrelevant columns

df = df.drop(['Admin1', 
              'SubDiv', 
              'Frmat', 
              'IM_Frmat', 
              'Deaths1', 
              'Deaths2',
              'Deaths3',
              'Deaths4',
              'Deaths5',
              'Deaths6',
              'Deaths7',
              'Deaths8',
              'Deaths9',
              'IM_Deaths1', 
              'IM_Deaths2', 
              'IM_Deaths3', 
              'IM_Deaths4'], 
               axis=1)
```

```python
df.head()
```

```python
df.head()
```

```python
# Use melt() to combine age features into one column, 'Age'
# Create new column, 'Freq'

df = pd.melt(df, 
             id_vars=['Country',
                      'Year',
                      'List',
                      'Cause',
                      'Sex'], 
             var_name='Age',
             value_name='Freq')
```

```python
df.head(10)
```

```python
print(df.shape)
```

```python
# There are many 0's in the Freq column, replace with nan and drop

df = df.replace(0, pd.np.nan).dropna()
```

```python
print(df.shape)
```

```python
# The value 'B00' refers to all causes, we don't need to distinguish between different diseases
# Select and drop everything that is not 'B00'

causes = df[df['Cause'] != 'B00'].index

df.drop(causes, inplace=True)
```

```python
print(df.shape)
```

```python
# Drop 'Country', 'Year', and 'Sex' columns
# Group by 'List', 'Cause', and 'Age'
# Aggregate 'Freq' column

df_1 = df.groupby(['List', 'Cause', 'Age'], as_index=False).agg({'Freq':'sum'})
```

```python
print(df_1.shape)
```

```python
df_1.head(10)
```

```python
df_1.tail(10)
```

```python
# Drop sparse classes

sparseClasses = df_1[df_1['Freq'] < 500000].index

df_1.drop(sparseClasses, inplace=True)

print(df_1.shape)
```

```python
chart = sns.barplot(x='Age', y='Freq', data=df_1, ci=None)

chart.set_xticklabels(chart.get_xticklabels(), rotation=45)

plt.show()
```

```python
# Save dataframe as a new .csv file

df_1.to_csv('cleaned_deaths_by_age.csv', index=None)
```
